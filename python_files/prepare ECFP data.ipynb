```python
import pandas as pd
import numpy as np
```

```python
def fold_ecfp_from_KNIME(file, length)->None:
    
    df = pd.read_csv(file)
    
    df.dropna(subset='Core (stucture) (#1) ECFP fingerprint', inplace=True)
    
    ecfp_list = []
    for i in df['Core (stucture) (#1) ECFP fingerprint']:
        
        unfolded = np.array(i[1:-1].split(',')).astype(np.int64)
        
        f = lambda x: abs(x) % length
        indices = f(unfolded)
        
        bitvector = np.zeros(length, dtype=int)
        
        bitvector[indices]=1
        ecfp_list.append(bitvector)
    
    output_df = pd.DataFrame(ecfp_list)
    output_df['SMILES'] = df['smiles(IREACTION)(SMILES)']
    output_df['Yield'] = df['yield']
    output_file = "{}_ECFP_{}.csv".format(file[:-4], length)
    
    output_df.to_csv(output_file, index=False)
```

```python
fold_ecfp_from_KNIME(r'CA_PA\amidation_carboxylicAcid_primaryAmine.csv', 2048)
```
